```python
import pandas #this module is used to fetch and analyse  data

data = pandas.read_csv("school.csv", parse_dates=['bday','enrolldate', 'expgradate']) # allows you to specify which columns in the CSV file contain date information.
data
```

```python
# Cleaning, Do we have empties? Do we have unwanted columns, Do we have unwanted row?
data.isnull().sum()
# Missing Data... Less than 2%, its Fine
```

```python
# Fill Empties/Imputation - Qualitative Data
data['Gender'].fillna(2.0, inplace = True)
data['Rank'].fillna(0.0, inplace = True)
data['Major'].fillna('Unknown', inplace = True)
data['Smoking'].fillna(3.0, inplace = True)
# TODO : Fill  -  state, LiveOnCampus, CommuteTime
#data['Smoking'].head(50)
#data.isnull().sum()
```

```python
data.head(35)
```

```python
# Replacing
data['Gender'].replace({0:'Male',
                        1:'Female',
                        2:'Neutral'},
                         inplace = True)

data['LiveOnCampus'].replace({0:'Off Campus',
                              1:'On Campus'},
                              inplace = True)

data['HowCommute'].replace({1:'Walk',
                              2:'Bike',
                              3:'Car',
                              4:'Public Transit',
                              5:'Other'},
                              inplace = True)
# TODO - Rank, Gender, other categorical
```

```python
# Dropping a whole column
# in pandas axis-1 refer to vertical/columns
# in pandas axis-0 refer to horizontal/rows
data.drop(['MileMinDur'], axis=1, inplace=True)
#data.isnull().sum()
data.shape
```

```python
# droping multiple duplicate rows
data.drop_duplicates()
data.shape
```

```python
#Drop rows
data = data[data['Gender'] !='Male']
data = data[data['SleepTime'] ==7]
# Filter for both Male and Female
# filtered_data = data[data['Gender'].isin(['Male', 'Female'])] #filters multiple categories in a single step
data
```

```python
# randomly select a sample of rows from the Dataset.
data.sample(n=1)
```

```python
# This is a method provided by pandas DataFrames for retrieving the rows with the largest values based on a specified column
data.nlargest(3, 'Weight')
# data.nsmallest(1, 'Math') #smallest value in a specific column.
```

```python
# Describing quantitative columns
data.describe()
```

```python
# Describing qualitative columns
# Average Math by Rank  -  Category 1, Continous - 2
data.groupby('Gender') [['Math','English']].mean() #grouping DataFrame rows based on one or more columns. 
```

```python
# Average Math by Rank  -  Category 3, Continous - 2
data.groupby(['State', 'Rank', 'Gender']) [['Math','English']].mean()
```

```python
# grouping DataFrame rows based on one or more columns. 
# It's a powerful tool used to perform various aggregation operations on groups created by groupby().  It allows you to calculate summary statistics or apply custom functions on each group efficiently.
data.groupby(['State', 'Rank', 'Gender']) [['Math','English']].agg(['sum', 'max', 'min', 'std', 'median'])
```

```python
# Apply Some function to combine columns and generate a new column with Totals - we use lambda
# a lambda function (often shortened to "lambda") is a concise way to define an anonymous function.
# add = lambda x, y: x + y
# result = add(5, 3)  # result will be 8
# assign() This is a method provided by pandas DataFrames for creating new columns while preserving the original DataFrame.
# Total  is the name of the new column you're adding to the DataFrame.
# lambda data is a lambda function that specifies how to calculate the values for the new column.
data = data.assign(Total=lambda new_data: data.Reading + data.Math + data.Writing)
data.head(4)
```

```python
data = data.assign(BMI=lambda data: data.Weight/(data.Height * data.Height))
data.head(5)
```

```python
# Lets create a function that coverts weight from pounds to kg
def convert(x):
    answer = 0.453592 * x
    return answer

data['Weight'] = data['Weight'].apply(convert)
data.head(5)
```

```python
def convert(x):
    answer = 0.3048 * x
    return answer

data['Height'] = data['Height'].apply(convert)
data.head(2)
```

```python
data.to_csv('cleaned.csv')
```
