```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv("school.csv" , parse_dates=['bday','enrolldate', 'expgradate'])
data_frame
```

```python
data_frame.shape
```

```python
data_frame.info()
```

```python
data_frame['Gender'].replace({0:'Male',
                        1:'Female',
                        2:'Neutral'},
                         inplace = True)

data_frame['LiveOnCampus'].replace({0:'Off Campus',
                              1:'On Campus'},
                              inplace = True)

data_frame['HowCommute'].replace({1:'Walk',
                              2:'Bike',
                              3:'Car',
                              4:'Public Transit',
                              5:'Other'},
                              inplace = True)
data_frame
```

```python
data_frame['Gender'].fillna('Male' , inplace=True)
data_frame['Writing'].fillna(77.68 , inplace=True)
```

```python
data_frame.info()
```

```python
# 1. What is the Average Study Time
# 2. What is the Average for Writing

data_frame[['StudyTime' , 'Writing']].mean()
```
