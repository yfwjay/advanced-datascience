```python
# Predict the amount a income one will get is based on their age and experience.
```

```python
array=income_dataframe.values
array
X=array[0:20,0:2] 
X
Y=array[0:20,2:4]
Y
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_test)
print("Predictions:",prediction)
```

```python
print("Expected outcome",Y_test)
```

```python
from sklearn.metrics import r2_score 
print(" The model is: ",r2_score(prediction,Y_test)*100  ," accurate")
```

```python
# 2.	How much will I earn as a 30 year old with 10 yearsâ€™ experience?

earnings = [[30 , 10 ]]
expected = model.predict(earnings)
print("Your expected earnings is : " , expected)
```

```python
customer_dataframe.info()
```

```python
array=customer_dataframe.values
array
X=array[0:660,0:9]
X
```

```python
from sklearn.cluster import KMeans as km
model=km(n_clusters=20,random_state=42)
model.fit(X)
```

```python
means = model.cluster_centers_
means
```

```python
clusters=pd.DataFrame(means ,columns=['Sl_No','Customer Key','Avg_Credit_Limit' , 'Total_Credit_Cards' , 'TT_Visits_bank' , 'TT_Visits_online' , 'TT_calls_made'] ) 
clusters
```

```python
customer_dataframe["group_number"]=model.labels_
customer_dataframe.head()
```
